```python
from json_tricks import load
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
all_res = load("experimental_results.json")
```

```python
all_res.keys()
```

```python
conformal_len = np.array([len(x) for x in all_res["conformal"]])
```

```python
topk_len = np.array([len(x) for x in all_res["topclass"]])
```

```python
fig, axn = plt.subplots(1, 2, sharex=True, sharey=False)
axn[0].hist(conformal_len)
axn[0].set_xlabel('Length of conformal set')
axn[0].set_ylabel('Frequency')
axn[1].hist(topk_len)
axn[1].set_xlabel('Length of tep-k set')
axn[1].set_ylabel('Frequency')
```

```python
all_two_class = np.where(conformal_len == 2)
```

```python
this_star = all_two_class[0][1]
```

```python
all_res['star'][this_star]
```

```python
spectrum = all_res['spectra'][this_star]
```

```python
padding = np.max(np.where(spectrum[0]==0)) + 1
```

```python
plt.plot(spectrum[0][padding:], spectrum[1][padding:])
plt.xlabel('Wavelength')
plt.ylabel('Flux')
```

```python
all_ten_class = np.where(conformal_len == 10)
```

```python
this_star = all_ten_class[0][0]
all_res['star'][this_star]
```
